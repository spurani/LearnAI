```python
# Import Libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import boston_housing
```

```python
# Load Boston Housing Data
(train_X, train_Y), (test_X, test_Y) = boston_housing.load_data()
```

```python
# Initialize/ Create Sequential Model
model = Sequential()
```

```python
# Add Dense Layer with input_shape as it is attached to input layer
model.add(Dense(1, input_shape=(train_X.shape[1],),activation='linear'))
```

```python
# Check the model summary
model.summary()
```

```python
# Model configuration
model.get_config()
```

```python
# Model compile
model.compile(optimizer='rmsprop', loss='mse', metrics=['mse','mae'])
```

```python
# Model train
model.fit(x=train_X, y=train_Y, batch_size=2, epochs=500, validation_split=0.2, shuffle=True)
```

```python
# Evaluate the model to check the metrics(mean squared error, mean absolute error) on test data
model.evaluate(test_X, test_Y)
```

```python
# Predict the model -> here the model will return the median price of the house in boston
y_pred=model.predict(test_X[:5])
```
