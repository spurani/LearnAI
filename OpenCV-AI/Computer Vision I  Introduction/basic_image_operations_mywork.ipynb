```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
im=cv2.imread("New_Zealand_Boat.jpg")
im=im[:,:,::-1]
plt.title("Original Image")
plt.imshow(im)
```

```python
im1=im.copy()
plt.title("Copied from original Image")
plt.imshow(im1)
```

```python
imatrix=np.zeros((100,200,3),dtype='uint8')
```

```python
print("imatrix size: ",imatrix.shape)
plt.imshow(imatrix)
```

```python
## First method
imatrix[:,:,:]=255
## Second method
#imatrix=imatrix*255
## Third method
#imatrix= 255 * np.ones((100,200,3),dtype='uint8')
print("imatrix size: ",imatrix.shape)
plt.imshow(imatrix)
```

```python
###Create image matrix of same size and datatype from image using ones_like() of numpy
imatrix1= 100 * np.ones_like(im)
print("imatrix size: ",imatrix1.shape)
plt.imshow(imatrix1)
```

```python
print(im[225:350,400:500])
```

```python
##Crop boat
cropim=im[225:350,400:500]
plt.imshow(cropim)
print("Cropped image size: ",cropim.shape)
```

```python
oricopy=im.copy()
plt.imshow(oricopy)
```

```python
oricopy[225:350,300:400]=cropim
oricopy[225:350,500:600]=cropim
```

```python
plt.figure(figsize=(10,10));
plt.subplot(121);plt.title("Cropped");plt.imshow(oricopy);
plt.subplot(122);plt.title("Original");plt.imshow(im);
```

```python
resizedboat=cv2.resize(im,(600,500),interpolation= cv2.INTER_LINEAR)
resizedboatdown=cv2.resize(im,(200,100),interpolation= cv2.INTER_LINEAR)
plt.figure(figsize=(10,10));
plt.subplot(131);plt.title("Resized Up");plt.imshow(resizedboat);
plt.subplot(132);plt.title("Resized Down");plt.imshow(resizedboatdown);
plt.subplot(133);plt.title("Original");plt.imshow(im);
```

```python
resizedboat=cv2.resize(im,None,fx=1.5,fy=1.5,interpolation= cv2.INTER_LINEAR);
resizedboatdown=cv2.resize(im,None,fx=0.5,fy=0.5,interpolation= cv2.INTER_LINEAR);
plt.figure(figsize=(15,15));
plt.subplot(131);plt.title("Resized Up");plt.imshow(resizedboat);plt.title(resizedboat.shape);
plt.subplot(132);plt.title("Resized Down");plt.imshow(resizedboatdown);plt.title(resizedboatdown.shape);
plt.subplot(133);plt.title("Original");plt.imshow(im);plt.title(im.shape);
```
