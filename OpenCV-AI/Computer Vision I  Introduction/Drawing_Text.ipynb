```python
import cv2
import numpy
from matplotlib import pyplot as plt
```

```python
image=cv2.imread("C:/Users/SP.000/Pictures/Saved Pictures/raptor.jpg")
image=image[:,:,::-1]
```

```python
plt.imshow(image)
```

```python
fontheight=50
thickness=2
text="F-22 Raptor heat burn"
fontscale=cv2.getFontScaleFromHeight(cv2.FONT_HERSHEY_SIMPLEX,fontheight,thickness)
print("FontScale: ",fontscale)
```

```python
textSize, baseLine = cv2.getTextSize(text,cv2.FONT_HERSHEY_SIMPLEX,fontscale,thickness)
print("TextSize: ",textSize)
print("BaseLine: ",baseLine)
```

```python
textWidth, textHeight=textSize
imageCopy=image.copy()
imageCopyHeight,imageCopyWidth=imageCopy.shape[:2]
print("TextWidth: ",textWidth)
print("TextHeight: ",textHeight)
print("ImageHeight: ",imageCopyHeight)
print("ImageWidth: ",imageCopyWidth)
```

```python
xcoordinate=(imageCopyWidth-textWidth)//2
ycoordinate=imageCopyHeight-textHeight
canvastopleft=(xcoordinate,ycoordinate-textHeight)
canvasbottomright=(xcoordinate+textWidth,ycoordinate+baseLine)
print("X coordinate: ",xcoordinate,"Ycoordinate: ",ycoordinate)
cv2.rectangle(imageCopy,canvastopleft,canvasbottomright,(255,255,255),-1,cv2.LINE_8)
lineleft=(xcoordinate,ycoordinate)
lineright=(xcoordinate+textWidth,ycoordinate)
cv2.line(imageCopy,lineleft,lineright,(0,255,255),3,cv2.LINE_AA)
cv2.putText(imageCopy,text,(xcoordinate,ycoordinate),cv2.FONT_HERSHEY_SIMPLEX,fontscale,(255,0,0),thickness,cv2.LINE_8)
plt.imshow(imageCopy)
```
