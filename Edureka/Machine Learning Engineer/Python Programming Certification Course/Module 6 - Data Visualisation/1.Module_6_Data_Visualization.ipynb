## multiline plots

```python
a = [1, 2, 3, 4]
plt.plot(a, [elem for elem in a])
plt.plot(a, [elem**2 for elem in a])
plt.plot(a, [elem**3 for elem in a])
plt.show()
```

```python
# using single method call
a = [1, 2, 3, 4]
# plt.plot(x1,y1, x2, y2, x3,y3)
plt.plot(a, [elem for elem in a],
         a, [elem**2 for elem in a],
         a, [elem**3 for elem in a])
plt.show()
```

```python
# adding Grids
a = [1, 2, 3, 4]
plt.plot(a, [elem for elem in a],
        a, [elem**2 for elem in a],
        a, [elem**3 for elem in a])
plt.grid(True)
plt.show()
```

```python
## limiting the axes
a = [1, 2, 3, 4]
plt.plot(a, [elem for elem in a],
        a, [elem**2 for elem in a],
        a, [elem**3 for elem in a])
plt.grid(True)

plt.xlim(0,2.5)  ### method 1
plt.ylim(-1,30) ####
plt.show()
```

```python
## limiting the axes
a = [1, 2, 3, 4]
plt.plot(a, [elem for elem in a],
        a, [elem**2 for elem in a],
        a, [elem**3 for elem in a])
plt.grid(True)

plt.axis([0, 2.5, 0, 30])  # method 2
plt.show()
```

```python
# adding labels to each curve
a = [1, 2, 3, 4]
plt.plot(a, [elem for elem in a], label='Linear')
plt.plot(a, [elem**2 for elem in a], label='square')
plt.plot(a, [elem**3 for elem in a], label='cube')

plt.legend()  #####
plt.show()
```

```python
#X axis Yaxis
# Title of the plot
a = [1, 2, 3, 4]
plt.plot(a, [elem for elem in a], label='Linear')
plt.plot(a, [elem**2 for elem in a], label='square')
plt.plot(a, [elem**3 for elem in a], label='cube')
plt.grid(True)
plt.xlabel('X-Axis')  ###
plt.ylabel('Y-Axis')  ###

plt.title("Polynomial Graph")

plt.legend()
plt.show()
```

```python
# saving plots
a = [1, 2, 3, 4]
plt.plot(a, [elem for elem in a], label='Linear')
plt.plot(a, [elem**2 for elem in a], label='square')
plt.plot(a, [elem**3 for elem in a], label='cube')
plt.grid(True)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("Polynomial Graph")
plt.legend()
plt.savefig('myplot_29thJune')  #####
plt.show()
```

## Bar Graph
- used to visually compare 2 or more value using rectangular bars

```python
import pandas as pd
data = pd.read_csv('AllCountries.csv')
data.head()
```

```python
# let's pick up first 10 countries
data_10 = data[:10]
plt.figure(figsize=(5,4))  # increase/control the size of the graph
plt.bar(data_10['Country'],
        data_10['LandArea'],
       )
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Land area")
plt.show()
```

## Pie-Chart

```python
country = ['India', 'US', 'China', 'Russia', 'Germany']
sales = [20000, 3000, 5000, 6000, 8000]
plt.pie(x=sales,
        labels=country)
plt.show()
```

```python
# showing percentage
country = ['India', 'US', 'China', 'Russia', 'Germany']
sales = [20000, 3000, 5000, 6000, 8000]
plt.pie(x=sales,
        labels=country,
        autopct='%.2f%%')  # f is for float value
plt.show()
```

```python
20000/(20000 + 22000)
```

```python
# showing percentage
country = ['India', 'US', 'China', 'Russia', 'Germany']
sales = [20000, 3000, 5000, 6000, 8000]
plt.pie(x=sales,
        labels=country,
        autopct='%d%%')  # d is for integer value
plt.show()
```

```python
# using explode
country = ['India', 'US', 'China', 'Russia', 'Germany']
sales = [20000, 3000, 5000, 6000, 8000]
plt.pie(x=sales,
        labels=country,
        autopct='%d%%',
        explode=(0, 1.5, 0, 0, 0.3))  # d is for whole number
plt.show()
```

## Scatter plot

```python
data = pd.read_csv('AllCountries.csv')
data.head()
```

```python
x = data['Population']
y = data['GDP']
plt.figure(figsize=(7,7))
plt.scatter(x,y)         ########
plt.xlabel("Population")
plt.ylabel("GDP")

plt.title("Population vs GDP")
plt.show()
```

```python
# limiting the axes to visualize better

x = data['Population']
y = data['GDP']
plt.figure(figsize=(7,7))
plt.scatter(x,y)

plt.xlim(0,100) #############
plt.xlabel("Population")
plt.ylabel("GDP")

plt.title("Population vs GDP")
plt.show()
```

## Formatting

```python
y = np.arange(1,4)
plt.plot(y, 'y')
plt.plot(y+5, 'm')
plt.plot(y+10, 'c')
plt.show()
```

```python
# control line styling
y = np.arange(1,30)
plt.plot(y,'--', 
        y*4, '-.',
        y*8, ':')
plt.show()
```

```python
# control marking styling
y = np.arange(1,10,0.8)
plt.plot(y, '*',
        y+0.5, 'o',
        y+ 1, 'D',
        y+2, '^',
        y+3, 's')
plt.show()
```
