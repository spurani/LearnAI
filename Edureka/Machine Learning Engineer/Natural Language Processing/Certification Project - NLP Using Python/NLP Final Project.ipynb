```python
# -*- coding: utf-8 -*-
"""
Created on Fri Jan 29 03:42:37 2021

@author: SP
"""

import os
import pandas as pd
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.tokenize import sent_tokenize, word_tokenize
from sklearn.cluster import KMeans
import numpy as np
from sklearn.metrics import silhouette_score

rootdir = 'BBC News Articles'
data_list = []
data = pd.DataFrame()
i = 1
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        file = open(os.path.join(subdir, file), encoding='utf-8')
        line = file.read().replace("\n","")
        print(subdir.rsplit('\\')[-1])
        data.set_value(i,'Article',line.lower())
        file.close()
        i+=1
```

```python
#data.to_csv("BBCNewsArticles.csv",index=False)
data['Articlenopunctuations'] = data['Article'].str.replace("[^\w|\d|\s]","")
#data['Articlenopunctuations'] = data['Article'].str.replace("[.,]","")
#data['Articlenopunctuations'] = data['Article'].str.replace("['']"," ")

stop = stopwords.words('english')
data['Articlepreprocessed']= data['Articlenopunctuations'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
data['Articletokens'] = data['Articlepreprocessed'].apply(word_tokenize)
# data.to_csv("BBCNewsArticles.csv",index=False)
all_words = []
for i in data['Articletokens']:
    for j in i:
        all_words.append(j)
corpus_tokens = list(set(all_words))
print(corpus_tokens)
```

```python
'''PresenceAbsence Vector'''
articlepavector = []

for j in data['Article']:
    vector = []
    for k in j:
        if(k in corpus_tokens):
            vector.append(1)
        else:
            vector.append(0)
    articlepavector.append(vector)
print(articlepavector)
# data.to_csv("BBCNewsArticles.csv",index=False)
```

```python
'''Count Vector'''
countvector = CountVectorizer()
X_counts = countvector.fit_transform(data['Articlepreprocessed'])
X_names = countvector.get_feature_names()
print(X_names)
print(len(X_counts))
print(X_counts[:,countvector.vocabulary_['sales']].toarray())
```

```python
'''TFID VECTOR'''
tf_vect = TfidfVectorizer(min_df=1,lowercase=True,stop_words='english')
tf_matrix = tf_vect.fit_transform(data['Articlepreprocessed'])
tf_names = tf_vect.get_feature_names()
print(tf_names)
print(type(tf_matrix))
print(tf_matrix.shape)
tf_df = pd.DataFrame(tf_matrix.toarray(),columns=tf_names)
print(tf_df)
```

```python
number_of_clusters=5
kmcountvector = KMeans(n_clusters=number_of_clusters)
kmcountvector.fit(X_counts)
label = kmcountvector.predict(X_counts)
print(label)
print(kmcountvector.labels_)
pd.crosstab(kmcountvector.labels_, data["Articlepreprocessed"]).to_csv("BBCNewsArticlesClustered.csv.csv")
print(f'Silhouette Score(n=5): {silhouette_score(X_counts, label)}')
```

```python
kmtfvect = KMeans(n_clusters=number_of_clusters)
kmtfvect.fit(tf_matrix)
labeltf = kmtfvect.predict(tf_matrix)
print(labeltf)
print(kmtfvect.labels_)
#pd.crosstab(kmtfvect.labels_, data["Articlepreprocessed"]).to_csv("BBCNewsArticlesClustered.csv.csv",'a')
```
