```python
def cfg_parse(sentence):
    sent_tk = pos_tag(word_tokenize(sentence))
    print(sent_tk)
    for pair in sent_tk:
        if pair[1]=='NNP':
            s_NP = "\'"+pair[0]+"\'"
        if pair[1]=='VBD' or pair[1]=='VBN':
            s_V = "\'"+pair[0]+"\'"
        if pair[1]=='NN':
            s_N = "\'"+pair[0]+"\'"
    
    cfg_grammar2 = CFG.fromstring("""
    S -> NP VP
    VP -> V N
    NP -> {}
    V -> {}
    N -> {}
    """.format(s_NP, s_V, s_N))
    print()
    for sentence in generate(cfg_grammar2):
        print(" ".join(sentence))
```
