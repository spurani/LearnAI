```python
import numpy as np
print(np.unique(trainy))
```

```python
print(trainX.dtype)
print(trainy.dtype)
print(testX.dtype)
print(testy.dtype)
print(trainX.shape)
print(trainy.shape)
print(testX.shape)
print(testy.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
trainX = trainX.astype('float32')
testX = testX.astype('float32')
trainX = trainX.reshape(60000,784)
testX = testX.reshape(10000,784)
```

```python
trainX /= 255
testX /= 255
```

```python
print(trainX)
print(testX)
```

```python
from tensorflow.keras.utils import to_categorical
trainy = to_categorical(trainy)
testy = to_categorical(testy)
```

```python
model = Sequential()
model.add(Dense(10,input_shape=(784,),activation="softmax",name="Dense_Layer"))
model.compile(optimizer="Adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
training = model.fit(trainX, trainy, batch_size=64, epochs=100, validation_split=0.2)
```

```python
test_loss, test_accuracy = model.evaluate(testX,testy)
```

```python
model.summary()
```

```python
model.get_config()
```

```python
from matplotlib import pyplot as plt
plt.plot(training.history['acc'])
plt.plot(training.history["val_acc"])
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.title("model accuracy")
plt.legend(["train","test"])
plt.show()
plt.plot(training.history['loss'])
plt.plot(training.history["val_loss"])
plt.xlabel("epoch")
plt.ylabel("loss")
plt.title("model loss")
plt.legend(["train","test"])
```

```python
##Add a hidden layer to the model to make it Multilayer Perceptron
model1 = Sequential()
N_hidden=128
model1.add(Dense(N_hidden,input_shape=(784,),name="Dense_Layer_1",activation="relu"))
model1.add(Dense(N_hidden,name="Dense_Layer_2",activation="relu"))
model1.add(Dense(10,name="Dense_Layer_3",activation="softmax"))
model1.compile(optimizer="Adam",metrics=["accuracy"],loss="categorical_crossentropy")
training1 = model1.fit(trainX,trainy,batch_size=64,epochs=100,validation_split=0.2)
model1.evaluate(testX,testy)
from matplotlib import pyplot as plt
plt.plot(training1.history["acc"])
plt.plot(training1.history["val_acc"])
plt.xlabel("accuracy")
plt.ylabel("epochs")
plt.title("Model Accuracy")
plt.legend(["train","test"])
plt.show()
plt.plot(training1.history["loss"])
plt.plot(training1.history["val_loss"])
plt.xlabel("loss")
plt.ylabel("epochs")
plt.title("Model Loss")
plt.legend(["train","test"])
```

```python
##Add Dropout to prevent overfitting and check its effect on accuracy
model2 = Sequential()
N_hidden=128
model2.add(Dense(N_hidden,input_shape=(784,),name="Dense_Layer_1",activation="relu"))
model2.add(Dropout(0.5))
model2.add(Dense(N_hidden,name="Dense_Layer_2",activation="relu"))
model2.add(Dropout(0.5))
model2.add(Dense(10,name="Dense_Layer_3",activation="softmax"))
model2.compile(optimizer="Adam",metrics=["accuracy"],loss="categorical_crossentropy")
training2 = model2.fit(trainX,trainy,batch_size=64,epochs=100,validation_split=0.2)
from matplotlib import pyplot as plt
plt.plot(training2.history["acc"])
plt.plot(training2.history["val_acc"])
plt.xlabel("accuracy")
plt.ylabel("epochs")
plt.title("Model Accuracy")
plt.legend(["train","test"])
plt.show()
plt.plot(training2.history["loss"])
plt.plot(training2.history["val_loss"])
plt.xlabel("loss")
plt.ylabel("epochs")
plt.title("Model Loss")
plt.legend(["train","test"])
model2.evaluate(testX,testy)
```
